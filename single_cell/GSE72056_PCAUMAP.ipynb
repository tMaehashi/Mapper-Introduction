# GSE72056

## 読み込み

```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
```

```python
data = pd.read_table('GSE72056_melanoma_single_cell_revised_v2.txt') 
data = data.set_index('Cell') # インデックスを遺伝子名に変更
data
# 遺伝子×細胞
# 23秒
```

```python
df = data.drop(['tumor', 'malignant(1=no,2=yes,0=unresolved)', 'non-malignant cell type (1=T,2=B,3=Macro.4=Endo.,5=CAF;6=NK)'], 
          axis=0) # 行（遺伝子）を削除
df
```

最初の3行を削除すると，論文のように23686の遺伝子になる．  
23686 * 4645

```python
malignant_label = np.array(data.iloc[1, :].to_list()) # 0から数えて1行目（悪性ラベル）を抽出
malignant_label
```

```python
tpm = df.values.T # 値のみ転置
tpm
```

行が細胞，列が遺伝子になった．

```python
tpm.shape
```

## 遺伝子の抽出

```python
tpm_mean = np.mean(tpm, axis=0) # 列ごと（遺伝子ごと）に平均発現量を計算
tpm_mean_sort = np.sort(tpm_mean)[::-1] # tpm_meanを大きい順にソート

tpm_var = np.var(tpm, axis=0) # 列ごと（遺伝子ごと）に発現量の分散を計算
tpm_var_sort = np.sort(tpm_var)[::-1] # tpm_varを大きい順にソート
```

平均／分散発現量が大きい方から7000個の遺伝子を抽出（配列の6999番まで）

```python
# 閾値
print(tpm_mean_sort[6999]) 
print(tpm_var_sort[6999])
```

```python
c = (tpm_mean >= tpm_mean_sort[6999]) & (tpm_var >= tpm_var_sort[6999])
G = tpm[:,c]
G.shape
```

## PCA，UMAP（２次元）

```python
from sklearn.decomposition import PCA
data_PCA = PCA(n_components=2).fit_transform(G)
```

```python
import umap
data_UMAP = umap.UMAP(n_components=2, random_state=0, n_neighbors=15).fit_transform(G)
```

## クラスタを見る

悪性ラベルで色付け  
0：不明，1：良性，2：悪性

```python
df_PCA = pd.DataFrame(data_PCA)
df_PCA['label'] = malignant_label
df_PCA
```

```python
colors = ['#5e4fa2', '#fee08b', '#9e0142']
labels = [0, 1, 2]
captions = ['不明', '良性', '悪性']
```

```python
import japanize_matplotlib # 日本語対応
for color, label, caption in zip(colors, labels, captions):
    tmp = df_PCA.loc[df_PCA['label'] == label]
    plt.scatter(tmp[0], tmp[1], color=color, label=caption)
plt.legend()
```

```python
df_UMAP = pd.DataFrame(data_UMAP)
df_UMAP['label'] = malignant_label
df_UMAP
```

```python
import japanize_matplotlib # 日本語対応
for color, label, caption in zip(colors, labels, captions):
    tmp = df_UMAP.loc[df_UMAP['label'] == label]
    plt.scatter(tmp[0], tmp[1], color=color, label=caption)
plt.legend()
```
