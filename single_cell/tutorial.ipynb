# Mapper チュートリアル

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np
```

## 読み込み・前処理
log2{FPKM+1}

```python
data = pd.read_csv('Trapnell.csv', index_col=0)
data = np.log2(data+1)
print(data.shape)
data.head()
```

## 遺伝子リスト

```python
# 重要遺伝子20個抽出
df = data.loc[:, ['ENSG00000081189', 'ENSG00000116604', 'ENSG00000111049', 'ENSG00000166825',\
               'ENSG00000134853', 'ENSG00000122180', 'ENSG00000140416', 'ENSG00000198467',\
               'ENSG00000125414', 'ENSG00000109063', 'ENSG00000149294', 'ENSG00000105048',\
               'ENSG00000118194', 'ENSG00000114854', 'ENSG00000170312', 'ENSG00000123374',\
               'ENSG00000134057', 'ENSG00000157456', 'ENSG00000110092', 'ENSG00000125968'
               ]]
df.head()
```

```python
genes = ['MEF2C', 'MEF2D', 'MYF5', 'ANPEP', 'PDGFRA', 'MYOG', 'TPM1', 'TPM2',\
         'MYH2', 'MYH3', 'NCAM1', 'TNNT1', 'TNNT2', 'TNNC1', 'CDK1', 'CDK2',\
         'CCNB1', 'CCNB2', 'CCND1', 'ID1']

df.columns = genes
df.head()
```

## 細胞サンプルの日付ラベルの取得

```python
day_label = [0 for i in range(96)]
for i in range(96):
    day_label.append(1)
for i in range(96):
    day_label.append(2)
for i in range(84):
    day_label.append(3)
len(day_label)
```

## 標準化とPCA

```python
# 遺伝子ごとに標準化
df_sc = (df - df.mean()) / df.std()
val = df_sc.values
df_sc.head()
# df_sc.var() # 分散１の確認
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2).fit_transform(val)

plt.axis('equal')
plt.scatter(pca[:, 0], pca[:, 1], c=day_label)
plt.title('PCA (2D)')
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=0, perplexity=20).fit_transform(val)

plt.axis('equal')
plt.scatter(tsne[:, 0], tsne[:, 1], c=day_label)
plt.title('t-SNE (2D)')
```

```python
import umap

UMAP = umap.UMAP(n_components=2, random_state=0, n_neighbors=15).fit_transform(val)

plt.axis('equal')
plt.scatter(UMAP[:, 0], UMAP[:, 1], c=day_label)
plt.title('UMAP (2D)')
```

## 標準化後の値の保存

```python
genes_dict = {}
for gene in genes:
    gene_list = df_sc[gene].to_list()  # 各列をリストに変換
    genes_dict[gene] = gene_list # 列名をキーとし、配列を値として辞書に追加
```

```python
import kmapper2 as km 
from kmapper2.plotlyviz import plotlyviz
import sklearn

mapper = km.KeplerMapper(verbose=0)
```

### mappertools
https://github.com/emerson-escolar/mappertools

```python
# !pip install mappertools-1.0.0.tar.gz
import mappertools.mapper.hierarchical_clustering as hc
```

フィルター関数・レンズ：PCA（２次元） ※井元先生推奨  
被覆の数：8  
被覆の重なり：50％ ※井元先生推奨  
クラスタリング：階層的クラスタリング（単連結法/最短距離法），１つ目のギャップ検出 ※井元先生推奨

```python
lens = pca
simplicial_complex = mapper.map(lens, X=val,
                                cover=km.Cover(n_cubes=8, perc_overlap=0.5),
                                clusterer=hc.HeuristicHierarchical(heuristic='firstgap'))
```

## Mapperグラフを出力
カラースケール：https://colorbrewer2.org/#type=diverging&scheme=Spectral&n=11

```python
none= [0 for i in range(df_sc.shape[0])]
_ = mapper.visualize(
        simplicial_complex,
        title='none',
        path_html='./output/none.html',
        color_values=none,
        color_function_name='none'
    )
```

### 細胞サンプルの取得日で色付け

```python
_ = mapper.visualize(
        simplicial_complex,
        title='day',
        path_html='./output/day.html',
        color_values=day_label,
        color_function_name='day_label'
    )
```

### 各遺伝子の発現量で色付け

```python
for gene in genes:
    _ = mapper.visualize(
        simplicial_complex,
        title=gene,
        path_html='./output/' + gene + '.html',
        color_values=genes_dict[gene],
        color_function_name=gene
    )
```
